Data consistency checks

```python
#checking for mixed data types
for col in df_temp.columns.tolist():
  weird = (df_temp[[col]].applymap(type) != df_temp[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_temp[weird]) > 0:
    print (col)
```

```python
#checking for missing values
pd.set_option('display.max_rows', None)
df_temp.isnull().sum()
```

```python
#Dropping Columns
#since the dataset pans across 59 years, I'm dropping some older year's data and including only recent 20 years data
df_temp1 = df_temp.drop(['Y1961', 'Y1962', 'Y1963', 'Y1964', 'Y1965', 'Y1966', 'Y1967', 'Y1968', 'Y1969', 'Y1970'], axis = 1)
df_temp1
```

```python
df_temp2 = df_temp1.drop(['Y1971', 'Y1972', 'Y1973', 'Y1974', 'Y1975', 'Y1976', 'Y1977', 'Y1978', 'Y1979', 'Y1980'], axis = 1)
```

```python
df_temp3 = df_temp2.drop(['Y1981', 'Y1982', 'Y1983', 'Y1984', 'Y1985', 'Y1986', 'Y1987', 'Y1988', 'Y1989', 'Y1990'], axis = 1)
```

```python
df_temp4 = df_temp3.drop(['Y1991', 'Y1992', 'Y1993', 'Y1994', 'Y1995', 'Y1996', 'Y1997', 'Y1998', 'Y1999', 'Y2000'], axis = 1)
df_temp4
```
