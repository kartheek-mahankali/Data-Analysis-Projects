## 4.6 Merging and exporting data - demo

##This script contains the following points:
##1. Create Data to experiment on
##2. Concatenate Data Frames
##3. Append data
##4. Merge data

```python
#Importing Libraries
import pandas as pd
import numpy as np
import os
```

##1. Create Data to experiment on

```python
# Define a dictionary containing January 2020 data
data1 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'purchased_meat':[0, 13, 3, 4],
    'purchased_alcohol':[1, 2, 10, 0],
    'purchased_snacks':[10, 5, 1, 7]}
```

```python
# Define a dictionary containing February 2020 data
data2 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
    'purchased_meat':[0, 10, 5, 3],
    'purchased_alcohol':[2, 4, 14, 0],
    'purchased_snacks':[15, 3, 2, 6]}
```

```python
#Convert the dictionary into dataframe
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
df_1 = pd.DataFrame(data2,index=[0, 1, 2, 3])
```

```python
df
```

```python
df_1
```

##2. Concatenate Data Frames

```python
#create a list that contains our data frames
frames = [df, df_1]
```

```python
frames
```

```python
# check the data types to be sure it is a list
type(frames)
```

```python
#concatenate the data frames with default option
df_concat = pd.concat(frames)
```

```python
#check the output
df_concat
```

```python
#concatenate dataframes using axis = 1, creating a wide format
df_concat = pd.concat(frames, axis = 1)
```

```python
# check the output
df_concat
```

##3. Append Data

```python
#appending data
df_appended = df.append(df_1)
```

```python
#check the output
df_appended
```

```python
# Create data with different columns from df
data3 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'days_purchased_on':[0, 13, 3, 4]}
```

```python
# Convert to dataframe
df_2 = pd.DataFrame(data3,index=[0, 1, 2, 3])
```

```python
df_2
```

```python
df_append_test = df.append(df_2)
```

```python
df_append_test
```

###4.Merging Data frames

```python
#Merge df and df_2 using customer_id as key
df_merged = df.merge(df_2, on = 'customer_id')
```

```python
#check the output
df_merged
```

```python
#merging df and df_2 with customer_id and month as keys
df_merged = df.merge(df_2, on = ['customer_id','month'])
```

```python
#check the output
df_merged
```

```python
#merge df and df_2 with customer_id and month as keys and add merge flag
df_merged = df.merge(df_2, on = ['customer_id', 'month'], indicator = True)
```

```python
df_merged
```

```python
df_merged['_merge'].value_counts()
```
