```python
#Step 6: Create a spending flag for each user based on the average price across all their orders 
#Aggregating user data with transform()
ords_prods_merge['Avg_amt_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check the output with the new column
ords_prods_merge.head()
```

```python
#Create a spending flag using the following criteria
#1st criteria: if the mean of the prices of products purchased by a user is lower than 10, then flag them as 'low spender'
ords_prods_merge.loc[ords_prods_merge['Avg_amt_spent'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# 2nd Criteria: if the mean of the prices of products purchased by a user is higher or equal to 10, then flag them as 'high spender'
ords_prods_merge.loc[(ords_prods_merge['Avg_amt_spent'] >= 10) , 'spending_flag'] = 'High spender'
```

```python
#check the output
ords_prods_merge[['user_id', 'Avg_amt_spent', 'spending_flag']].head()
```

```python
#check the count of spending flag
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#Step 7: create an order frequency flag that marks the regularity of a user's ordering behaviour 
#Aggregating user's, 'days_since_prior_order' using transform()
ords_prods_merge['frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge['frequency'].dtype
```

```python
ords_prods_merge['frequency'] = ords_prods_merge['frequency'].astype('int64')
```

```python
#check the output
pd.set_option('display.max_columns', None)
ords_prods_merge.head()
```

```python
#Create a frequency flag
#1st criteria: if the median of 'days_since_prior_order' is higher than 20, then the customer should be labeled 'non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['frequency'] > 20, 'frequency_flag'] = 'Non-Frequent customer'
```

```python
#2nd Criteria: if the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a 'regular customer'
ords_prods_merge.loc[(ords_prods_merge['frequency'] > 10) & (ords_prods_merge['frequency'] <= 20), 'frequency_flag'] = 'Regular Customer'
```

```python
#3rd Criteria: if the median is lower than or equal to 10, then the customer should be labeled a 'Frequent Customer'
ords_prods_merge.loc[ords_prods_merge['frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#check the output
ords_prods_merge[['user_id', 'frequency', 'frequency_flag']].head()
```

```python
#check the output with value_counts
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
